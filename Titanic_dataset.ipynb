```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Load Dataset
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
# 2. Data Cleaning: Handling missing values in 'Age'
# Missing ages ko median (beech ki value) se bhar rahe hain
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# 3. Data Analysis
print("--- Survival Analysis Results ---")
```

```python
# Question: Who survived more?
gender_survival = df.groupby('Sex')['Survived'].mean() * 100
print(f"\n1. Survival by Gender:\n{gender_survival}")
```

```python
# Question: Did class affect survival?
class_survival = df.groupby('Pclass')['Survived'].mean() * 100
print(f"\n2. Survival by Class:\n{class_survival}")
```

```python
# Question: Survival rate by age group
bins = [0, 18, 35, 60, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
age_survival = df.groupby('AgeGroup')['Survived'].mean() * 100
print(f"\n3. Survival by Age Group:\n{age_survival}")
```

```python
# 4. Visualizations
plt.figure(figsize=(15, 5))
```

```python
# Plot 1: Survival by Gender
plt.subplot(1, 3, 1)
sns.barplot(x='Sex', y='Survived', data=df, palette='viridis')
plt.title('Survival Rate by Gender')
```

```python
# Plot 2: Survival by Class
plt.subplot(1, 3, 2)
sns.barplot(x='Pclass', y='Survived', data=df, palette='magma')
plt.title('Survival Rate by Class')
```

```python
# Plot 3: Histogram of Passenger Ages
plt.subplot(1, 3, 3)
sns.histplot(df['Age'], bins=20, kde=True, color='blue')
plt.title('Passenger Age Distribution')
```

```python
plt.tight_layout()
plt.show()
```
